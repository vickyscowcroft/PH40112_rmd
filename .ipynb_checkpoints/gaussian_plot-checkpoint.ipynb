```python
%config InlineBackend.figure_format = 'retina'
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

import matplotlib
font = {'size'   : 16}

matplotlib.rc('font', **font)
```

```python
mu = 0
sigma = 1
x1 = -8
x2 = 8
```

```python
z1 = ( x1 - mu ) / sigma
z2 = ( x2 - mu ) / sigma
```

```python
x = np.arange(z1, z2, 0.001) # range of x in spec
x_all = np.arange(-10, 10, 0.001) # entire range of x, both in and out of spec
# mean = 0, stddev = 1, since Z-transform was calculated
y = norm.pdf(x,0,1)
y2 = norm.pdf(x_all,0,1)
```

```python
x_1sig = np.arange(-1, 1, 0.001)
x_2sig = np.arange(-2, 2, 0.001)
x_3sig = np.arange(-3, 3, 0.001)

y_1sig = norm.pdf(x_1sig,0,1)
y_2sig = norm.pdf(x_2sig,0,1)
y_3sig = norm.pdf(x_3sig,0,1)

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
plt.tight_layout()

#plt.style.use('fivethirtyeight')
ax.plot(x_all,gauss(x_all), color='grey')

colors = ['#ff9999','#66b3ff','#FFD700', '#99ff99']
## red, blue, gold, green, 
#ax.fill_between(x,y,0,alpha=0.3, color='b') 0, 2, 1, 3
ax.fill_between(x_all,y2,0, color=colors[0], alpha=0.8)
ax.fill_between(x_3sig,y_3sig,0,color=colors[1], alpha=0.3)
ax.fill_between(x_2sig,y_2sig,0, color=colors[1], alpha=0.3)
ax.fill_between(x_1sig,y_1sig,0, color=colors[1], alpha=0.4)
ax.vlines(x=-1,ymin=0,ymax=gauss(-1.),linestyle='dashed',color='k')
ax.vlines(x=1,ymin=0,ymax=gauss(1.),linestyle='dashed',color='k')
ax.vlines(x=-2,ymin=0,ymax=gauss(-2.),linestyle='dashed',color='k')
ax.vlines(x=2,ymin=0,ymax=gauss(2.),linestyle='dashed',color='k')
ax.vlines(x=-3,ymin=0,ymax=gauss(-3.),linestyle='dashed',color='k')
ax.vlines(x=3,ymin=0,ymax=gauss(3.),linestyle='dashed',color='k')

ax.set_xlim([-5,5])
ax.set_xlabel('No. of Standard Deviations Outside the Mean')
#ax.set_yticklabels([])
#ax.set_title('Normal Gaussian Curve')

sig1_tx = ('$\pm 1 \sigma = 68.3$%')
sig2_tx = ('$\pm 2 \sigma = 95.4$%')
sig3_tx = ('$\pm 3 \sigma = 99.7$%')

ax.annotate(sig1_tx,xy=(-2, 0.3), xycoords='data', ha='center')
ax.annotate("", xy=(-0.1, 0.24), xytext=(-1.5,0.29), xycoords='data', ha='center', arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=(-1, 0.23), xytext=(1, 0.23), xycoords='data', arrowprops=dict(arrowstyle="<|-|>"))

ax.annotate(sig2_tx,xy=(-3, 0.1), xycoords='data', ha='center')
ax.annotate("", xy=(-1.3, 0.05), xytext=(-2.5,0.09), xycoords='data', ha='center', arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=(-2, 0.04), xytext=(2, 0.04), xycoords='data', arrowprops=dict(arrowstyle="<|-|>"))


ax.annotate(sig3_tx,xy=(3.8, 0.07), xycoords='data', ha='center')
ax.annotate("", xy=(2.5, 0.005), xytext=(3.5,0.06), xycoords='data', ha='center', arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=(-3, 0.005), xytext=(3, 0.005), xycoords='data', arrowprops=dict(arrowstyle="<|-|>"))

plt.savefig('normal_curve.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
def gauss(x):
    gauss = 1./(np.sqrt(2*np.pi)) * np.exp((-1./2.)*(x)**2)
    return(gauss)
```
